````python
import pandas as pd
import unsloth
from llama_cpp import Llama
# train_dataset을 불러오고 싶은 경우
from datasets import load_from_disk
dataset = load_from_disk("./data/train_dataset/")
print(dataset['train'][0])
# df_train = dataset['train'].to_pandas().to_csv('/data/ephemeral/home/my-project/data/df_train.csv',index=False)
base_dir ='/data/ephemeral/nlp_workspace/data/'
data_dir ='df_train.csv'
df = pd.read_csv(f"{base_dir}{data_dir}")

llm = Llama.from_pretrained(
    repo_id="unsloth/Qwen3-30B-A3B-Instruct-2507-GGUF",
    filename="Qwen3-30B-A3B-Instruct-2507-UD-Q6_K_XL.gguf",
    n_ctx=32768,
    n_gpu_layers=-1,
    flash_attn=True,
    verbose=False
)


'''
위 코드로 모델 저장 이후에는 아래 주석된 llm사용 다만 llama.cpp-server 형식이 훨씬 편할것 같음
'''

# llm = Llama(
#     model_path="/data/ephemeral/home/.cache/huggingface/hub/models--unsloth--Qwen3-30B-A3B-Instruct-2507-GGUF/snapshots/eea7b2be5805a5f151f8847ede8e5f9a9284bf77/Qwen3-30B-A3B-Instruct-2507-UD-Q5_K_XL.gguf",
#     n_ctx=32768,
#     n_gpu_layers=-1,
#     flash_attn=True,
#     verbose=False
# )

system_msg = """
당신은 주어진 텍스트에서 사용자의 질문에 대한 정답을 정확하게 추출하는 '정보 추출기(Information Extractor)'입니다.
추론 과정 이후에는 질문이 요구한 정답을 정확한 형태로 기재할 준비가 되었는지 다시 한 번 검토한 다음, 맨 마지막 행에 오직 아래의 JSON 포맷으로만 출력하십시오.
{"extracted_answer": "정답"}
"""

for idx, row in df.iterrows():
    print(idx)

    user_content = f"텍스트\n{row['title']}\n{row['context']}\n\n질문\n{row['question']}"
    messages = [
        {"role": "system", "content": system_msg},
        {"role": "user", "content": user_content}
    ]

    s = 0
    f = 0
    while (s + f) < 5:
        out = llm.create_chat_completion(
            messages=messages,
            max_tokens=8192,
            temperature=1.5,
        )

        if out["choices"][0]["message"]["content"].count('"') > 1 and out["choices"][0]["message"]["content"].split('"')[-2].strip() in row['context']:
            df.loc[idx, f'response_{s}'] = out["choices"][0]["message"]["content"]
            print(out["choices"][0]["message"]["content"].split('"')[-2].strip())
            s += 1
        else:
            f += 1
        llm.reset()

    df.to_csv(f"{base_dir}Qwen3.csv", index=False)
````
